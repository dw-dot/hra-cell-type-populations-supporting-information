## Ginty - Skin

```python
centroids_csvs = glob.glob("/Users/abhiroop/Developer/cns/skin_12_data/*/centroids.csv")
region_to_hubmap = pd.read_csv("/Users/abhiroop/Developer/cns/Ginty_skin_donor_to_hubmap.csv")
```

```python
region_to_hubmap
```

```python
centroids_csvs
```

```python
master_df = pd.DataFrame(columns=['Name','Skin',
 'CD31',
 'T-Helper',
 'DDB2',
 'KI67',
 'P53',
 'CD68',
 'T-Reg',
 'T-Killer', 'Total'
])

for centroid_csv_path in centroids_csvs:
    name = centroid_csv_path.split('/')[-2].replace('_', ' ').replace('r', 'R')
    df = pd.read_csv(centroid_csv_path)
    try:
        name = region_to_hubmap[region_to_hubmap['Region'] == name]['hubmap id'].values[0]
    except KeyError:
        pass
    row = {
        'Name': name
    }
    cell_counts = df['cell_type'].value_counts()
    cell_counts['Total'] = sum(cell_counts.values)
    row.update(cell_counts)
    master_df = master_df.append(row, ignore_index=True)

```

## Hickey - Intestine

```python
import pandas as pd

master_df2 = pd.DataFrame(columns=[
    'Name', 'NK', 'Enterocyte', 'MUC1+ Enterocyte', 'TA', 'CD66+ Enterocyte',
       'Paneth', 'Smooth muscle', 'M1 Macrophage', 'Goblet',
       'Neuroendocrine', 'CD57+ Enterocyte', 'Lymphatic', 'CD8+ T', 'DC',
       'M2 Macrophage', 'B', 'Neutrophil', 'Endothelial', 'Cycling TA',
       'Plasma', 'CD4+ T cell', 'Stroma', 'Nerve', 'ICC', 'CD7+ Immune', 'Total'
])

cell_csv = pd.read_csv("/Users/abhiroop/Developer/cns/CODEX_HuBMAP_alldata_Dryad.csv")

for region in list(cell_csv['unique_region'].value_counts().keys()):
    region_csv = cell_csv[cell_csv['unique_region'] == region]

    row = {
        'Name': region
    }
    cell_counts = region_csv['Cell Type'].value_counts()
    cell_counts['Total'] = sum(cell_counts.values)
    row.update(cell_counts)
    master_df2 = master_df2.append(row, ignore_index=True)
```

```python
conversion_csv = pd.read_csv('/Users/abhiroop/Developer/cns/cell_count_compute/hickey_stanford_to_hubmap.csv')
```

```python
conversion_csv.dropna(inplace=True)
```

```python
hubmap_ids = []

for name in master_df2['Name']:
    donor, region = name.split('_')
    donor_conversion = conversion_csv.loc[conversion_csv['Stanford Donor'].str.contains(donor, case=False)]
    donor_region_conversion = donor_conversion.loc[donor_conversion['Region Tracker'].str.contains(region, case=False)]
    hubmap_id = list(donor_region_conversion['HuBMAP ID'].values)
    if len(hubmap_id) > 1:
        donor_region_conversion = donor_conversion[donor_conversion['Region Tracker'] == region]
        new_hubmap_id = list(donor_region_conversion['HuBMAP ID'].values)
        print(new_hubmap_id)
        if len(new_hubmap_id) == 0:
            pass 
        if len(new_hubmap_id) == 1:
            hubmap_id = new_hubmap_id

    if len(hubmap_id) > 1:
        hubmap_ids.append(','.join(hubmap_id))
    elif len(hubmap_id) == 1:
        hubmap_ids.append(hubmap_id[0])
    else:
        hubmap_ids.append('')
```

## Angelo - Maternal-Fetal Interface

```python
import pandas as pd

master_df3 = pd.DataFrame(columns=[
    'Name',
])

cell_csv = pd.read_csv("/Users/abhiroop/Developer/cns/neighborhood_scripts_sharing/Supplementary_table_3_single_cells_updated.csv")

for region in list(cell_csv['Point'].value_counts().keys()):
    region_csv = cell_csv[cell_csv['Point'] == region]

    row = {
        'Name': region
    }
    cell_counts = region_csv['lineage'].value_counts()
    cell_counts['Total'] = sum(cell_counts.values)
    row.update(cell_counts)
    master_df3 = master_df3.append(row, ignore_index=True)
```

```python
for col in master_df3.columns[1:]:
    master_df3[col] = master_df3[col].astype('Int64')
```
